```python
import os
import numpy as np
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# --- Dataset Loader ---
class CarDataset(Dataset):
    def __init__(self, pos_dir, neg_dir, image_size=(64, 64)):
        self.data = []
        self.labels = []
        self.image_size = image_size
        self.load_images(pos_dir, 1)
        self.load_images(neg_dir, 0)

    def load_images(self, folder, label):
        for fname in os.listdir(folder):
            if fname.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp')):
                try:
                    path = os.path.join(folder, fname)
                    img = Image.open(path).convert('L').resize(self.image_size)
                    img = np.array(img, dtype=np.float32) / 255.0
                    self.data.append(img[np.newaxis, :, :])  # [1, H, W]
                    self.labels.append(label)
                except:
                    continue

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return torch.tensor(self.data[idx]), torch.tensor(self.labels[idx])

# --- LeNet-5 Architecture ---
class LeNet(nn.Module):
    def __init__(self):
        super(LeNet, self).__init__()
        self.model = nn.Sequential(
            nn.Conv2d(1, 6, kernel_size=5),      # (64 - 5 + 1) = 60 → 6×60×60
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),  # → 6×30×30
            nn.Conv2d(6, 16, kernel_size=5),     # → 16×26×26
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),  # → 16×13×13
            nn.Flatten(),
            nn.Linear(16 * 13 * 13, 120),
            nn.Tanh(),
            nn.Linear(120, 84),
            nn.Tanh(),
            nn.Linear(84, 2)  # 2 classes: pos/neg
        )

    def forward(self, x):
        return self.model(x)

# --- Paths (Update if needed) ---
pos_path = r"C:\Users\Adil\Downloads\objects_2011_a\objects_2011_b\labeldata\pos"
neg_path = r"C:\Users\Adil\Downloads\objects_2011_a\objects_2011_b\labeldata\neg"

# --- Prepare Dataset ---
dataset = CarDataset(pos_path, neg_path)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_ds, val_ds = random_split(dataset, [train_size, val_size])
train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=32)

# --- Training Setup ---
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = LeNet().to(device)
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# --- Training Loop ---
train_losses, val_losses, val_accuracies = [], [], []
epochs = 10

for epoch in range(epochs):
    model.train()
    train_loss = 0
    for X, y in train_loader:
        X, y = X.to(device), y.to(device)
        optimizer.zero_grad()
        output = model(X)
        loss = loss_fn(output, y)
        loss.backward()
        optimizer.step()
        train_loss += loss.item()
    train_losses.append(train_loss / len(train_loader))

    # Validation
    model.eval()
    val_loss, correct, total = 0, 0, 0
    y_true, y_pred = [], []
    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(device), y.to(device)
            output = model(X)
            loss = loss_fn(output, y)
            val_loss += loss.item()
            preds = torch.argmax(output, dim=1)
            y_true.extend(y.cpu().numpy())
            y_pred.extend(preds.cpu().numpy())
            correct += (preds == y).sum().item()
            total += y.size(0)
    val_losses.append(val_loss / len(val_loader))
    val_accuracies.append(correct / total)
    print(f"Epoch {epoch+1}: Train Loss={train_losses[-1]:.4f}, Val Acc={val_accuracies[-1]:.4f}")

# --- Evaluation ---
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=["neg", "pos"]))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
ConfusionMatrixDisplay(cm, display_labels=["neg", "pos"]).plot(cmap="Blues")
plt.title("Confusion Matrix")
plt.show()

# --- Plot Loss and Accuracy ---
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(train_losses, label='Train Loss')
plt.plot(val_losses, label='Val Loss')
plt.legend()
plt.title("Loss per Epoch")
plt.subplot(1, 2, 2)
plt.plot(val_accuracies, label='Val Accuracy')
plt.legend()
plt.title("Validation Accuracy")
plt.tight_layout()
plt.show()
```
